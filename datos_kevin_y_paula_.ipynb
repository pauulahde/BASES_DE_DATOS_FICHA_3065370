```python
# ===========================================================
# 1. Importar librerías
# ===========================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ===========================================================
# 2. Crear datos del ejercicio
# ===========================================================
datos = {
    "Mes": [
        "Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
        "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"
    ],

    "Portatiles": [20, 18, 25, 30, 28, 35, 40, 38, 36, 34, 32, 30],
    "Monitores":  [15, 16, 14, 18, 20, 22, 24, 23, 21, 19, 18, 17],
    "Teclados":   [40, 42, 38, 45, 47, 50, 55, 53, 52, 48, 46, 44]
}

df = pd.DataFrame(datos)

print("=== DataFrame cargado ===")
print(df)

# ===========================================================
# 3. Exploración básica
# ===========================================================
print("\n=== Primeras filas ===")
print(df.head())

print("\n=== Información general ===")
print(df.info())

print("\n=== Estadísticas descriptivas ===")
print(df.describe())

# ===========================================================
# 4. Estadísticas con NumPy
# ===========================================================
print("\n=== Estadísticas con NumPy ===")

# Portátiles
ventas_port = df["Portatiles"].values
print("\nPortátiles:")
print("Media:", np.mean(ventas_port))
print("Mediana:", np.median(ventas_port))
print("Desviación estándar:", np.std(ventas_port))

# Monitores
ventas_mon = df["Monitores"].values
print("\nMonitores:")
print("Media:", np.mean(ventas_mon))
print("Mediana:", np.median(ventas_mon))
print("Desviación estándar:", np.std(ventas_mon))

# Teclados
ventas_tec = df["Teclados"].values
print("\nTeclados:")
print("Media:", np.mean(ventas_tec))
print("Mediana:", np.median(ventas_tec))
print("Desviación estándar:", np.std(ventas_tec))

# ===========================================================
# 5. Nueva columna: Total Mensual
# ===========================================================
df["Total_Mensual"] = df["Portatiles"] + df["Monitores"] + df["Teclados"]

print("\n=== DataFrame con Total Mensual ===")
print(df)

# ===========================================================
# 6. Mes con mayores ventas totales
# ===========================================================
mes_max = df.loc[df["Total_Mensual"].idxmax(), "Mes"]
max_ventas = df["Total_Mensual"].max()

print("\nMes con mayores ventas totales:")
print(f"{mes_max} con {max_ventas} unidades")

# ===========================================================
# 7. Gráficos
# ===========================================================

# ---- 7.1 Gráfico de líneas por producto ----
plt.figure(figsize=(10, 5))
plt.plot(df["Mes"], df["Portatiles"], marker="o", label="Portátiles")
plt.plot(df["Mes"], df["Monitores"], marker="o", label="Monitores")
plt.plot(df["Mes"], df["Teclados"], marker="o", label="Teclados")

plt.title("Ventas mensuales por producto - TecnoPlus")
plt.xlabel("Mes")
plt.ylabel("Unidades vendidas")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# ---- 7.2 Gráfico de barras del total mensual ----
plt.figure(figsize=(10, 5))
plt.bar(df["Mes"], df["Total_Mensual"], color='skyblue')

plt.title("Ventas totales mensuales - TecnoPlus")
plt.xlabel("Mes")
plt.ylabel("Unidades totales vendidas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ===========================================================
# 8. Conclusiones
# ===========================================================
print("\n=== Conclusiones del análisis ===")
print("1. Tendencias: Las ventas tienden a aumentar hacia mitad de año, con un pico en julio/agosto.")
print("2. Producto más fuerte: Los teclados son el producto con más ventas a lo largo de todo el año.")
print("3. Mes pico: El mes con más ventas totales es:", mes_max)

```
